```python
import re

def convert_to_html(text):
    # Split the text into paragraphs based on empty lines
    paragraphs = text.strip().split('\n\n')
    
    html_output = ''
    
    paragraph_i = 0
    for paragraph in paragraphs:
        # Check for italic and bold formatting using regular expressions
        paragraph = re.sub(r'__(.*?)__', r'<i>\1</i>', paragraph)
        paragraph = re.sub(r'\*\*(.*?)\*\*', r'<b>\1</b>', paragraph)
        
        # Wrap the paragraph in a <p> tag
        paragraph_html = f'<p id="essay-paragraph-{paragraph_i}" class="essay">{paragraph}</p>\n'

        # Split the paragraph_html into lines shorter than 80 characters
        lines = []
        line = ''
        for char in paragraph_html:
            if char == '\n':
                lines.append(line)
                line = ''
            else:
                line += char
                if len(line) >= 80 and line[-1] == ' ':
                    lines.append(line[:-1])
                    line = ''
        lines.append(line)

        # Add the lines to the html_output
        html_output += '\n'.join(lines)
        
        # Add an empty line after each paragraph
        html_output += '\n'

        paragraph_i += 1
    
    return html_output

text = """
I’m good at a lot of things. I’m not very good at most things. That’s the case for most people. Even if you think that you’re good at most things, you’re probably not. You’re not good at badminton and cooking and crocheting and calculus and making jokes and writing and small talk and asking questions and answering questions and __and and__ (those are all things that I’m not nearly as good at as I want to be). I’ve taken some steps to be better at some of those things. No, I am not making unnecessary small talk with anyone. What you’re reading right now is my attempt at getting better at writing.

I read somewhere something along the lines of “successful people do ordinary things for an extraordinary amount of time” - maybe it’s a Woodie Guthrie or Bob Dylan or Michael Jackson or Winston Churchill line.  I don’t want to be particularly successful—just want to get better at writing for now. This is my collection of short essays, and just things I’ve written.

Starting this week (first week of March, 2024), I’m gonna publish one piece of writing, like an essay, on this website. I might share it on other platforms or mediums, I don’t know yet. I know most of these essays will get very few, or no, readers, but that’s OK. For once, I am creating for myself, and for the sake of getting better at something.

Every Sunday, at least for the foreseeable future, you can expect to find an essay from me on here. I am setting myself no rules regarding the length or the content of the essay, but I’ll aim to make it about one page, and I’ll aim to vary the content. But every Sunday, let’s say at 2pm (local time wherever I am), you can expect to find one essay from me on here.

There is an entirely arbitrary anonymous “like” button on each essay. Feel free to click it; it doesn’t do anything other than just makes me feel good. It doesn’t feed any algorithm, give you any recommendations, or do anything else that’s annoying (or functional). The share button on each essay does work; it does exactly what it says on the tin. Feel free to use it if you think someone else may be interested in that specific essay. The comment button is an entirely anonymous way for you to share any thoughts you have. I would really appreciate a comment from you. Comments could be regarding the content, your thoughts, or my writing style (or mistakes)—they are completely anonymous (there is no way for me to know who you are), and only I can see them. I don’t plan on sharing the comments publicly, although I might mention them in future essays.

Thank you for taking a moment or two to read something I’ve written as part of this __silly__ little project. 
"""

html_output = convert_to_html(text)
print(html_output)
```
